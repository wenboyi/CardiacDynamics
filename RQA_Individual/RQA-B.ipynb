```python
import pandas as pd
import numpy as np
import os
from teaspoon.parameter_selection.FNN_n import FNN_n
from teaspoon.parameter_selection.MI_delay import MI_for_delay
from numpy import genfromtxt
from pyrqa.time_series import TimeSeries
from pyrqa.settings import Settings
from pyrqa.analysis_type import Classic
from pyrqa.neighbourhood import FixedRadius
from pyrqa.metric import EuclideanMetric
from pyrqa.computation import RQAComputation
from sklearn.preprocessing import StandardScaler
from scipy import stats
```

```python
filtered_files=[]
for file in os.listdir('/Users/wenboyi/Desktop/HR/Indi-RQA/RRB/'):
    if file.endswith('.' + 'csv'):
        filtered_files.append(file)
print(len(filtered_files))
```

```python
def estimation(time_series):
    tau = MI_for_delay(time_series, plotting = False, method = 'basic', h_method = 'sturge', k = 2, ranking = True)
    perc_FNN, n = FNN_n(time_series, tau, plotting = False)
    return tau,n
```

```python
rl=[]
det=[]
lam=[]
rr=[]
fl=[]
delayl,diml=[],[]
for filename in filtered_files:
    ts=[]
    df=genfromtxt('./RRB/'+filename,delimiter=',')
    ts=df.astype(np.float)
    ts=stats.zscore(ts)
    delay,dim = estimation(ts)
    delayl.append(delay)
    diml.append(dim)
    for i in range(500):
        r=np.arange(0,5,0.01)
        #print(r.shape)
        time_series = TimeSeries(ts,
                             embedding_dimension=dim,
                             time_delay=delay)
        settings = Settings(time_series,
                        analysis_type=Classic,
                        neighbourhood=FixedRadius(r[i]),
                        similarity_measure=EuclideanMetric,
                        theiler_corrector=1)
        computation = RQAComputation.create(settings,verbose=True)
        result = computation.run()
        result.min_diagonal_line_length = 2
        result.min_vertical_line_length = 2
        result.min_white_vertical_line_length = 2
        if abs(result.recurrence_rate - 0.055) <= 0.005:
            fl.append(filename)
            rl.append(r[i])
            rr.append(result.recurrence_rate)
            lam.append(result.laminarity)
            det.append(result.determinism)
            from pyrqa.computation import RPComputation
            from pyrqa.image_generator import ImageGenerator
            computation = RPComputation.create(settings)
            result = computation.run()
            ImageGenerator.save_recurrence_plot(result.recurrence_matrix_reverse,filename+'_B.png')
            break
            break
        elif abs(result.recurrence_rate - 0.053) > 0.005 and i == 499:
            fl.append(filename+'invalid')
            rl.append('invalid')
            rr.append('invalid')
            lam.append('invalid')
            det.append('invalid')
            break
```

```python
newcsv = pd.DataFrame(fl, columns=['filename'])
#newcsv['Delay'] = delayl
newcsv['Dim']=diml
newcsv['Radiaus'] = rl
newcsv['DET'] = det
newcsv['LAM'] = lam
newcsv['RR'] = rr
```
